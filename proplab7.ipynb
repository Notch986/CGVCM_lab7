```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Cargar imágenes
img1 = cv2.imread("coraje.jpeg")
img2 = cv2.imread("vizcarra.jpg")
img3 = cv2.imread("perro.jpeg")
```

```python
# Obtener las dimensiones de las imágenes
h1, w1 = img1.shape[:2]
h2, w2 = img2.shape[:2]
h3, w3 = img3.shape[:2]

# Imprimir los valores
print(f"Dimensiones de la imagen 1: altura = {h1}, ancho = {w1}")
print(f"Dimensiones de la imagen 2: altura = {h2}, ancho = {w2}")
print(f"Dimensiones de la imagen 3: altura = {h3}, ancho = {w3}")

# Encontrar la imagen más grande
max_height = max(h1, h2, h3)
max_width = max(w1, w2, w3)

print(f"altura maxima = {max_height}, ancho maximo = {max_width}")

# Redimensionar las imágenes
img1_resized = cv2.resize(img1, (max_width, max_height))
img2_resized = cv2.resize(img2, (max_width, max_height))
img3_resized = cv2.resize(img3, (max_width, max_height))

# Convertir las imágenes de BGR a RGB para matplotlib
img1_resized_rgb = cv2.cvtColor(img1_resized, cv2.COLOR_BGR2RGB)
img2_resized_rgb = cv2.cvtColor(img2_resized, cv2.COLOR_BGR2RGB)
img3_resized_rgb = cv2.cvtColor(img3_resized, cv2.COLOR_BGR2RGB)

# Crear una figura y ejes
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Mostrar las imágenes en los ejes
axes[0].imshow(img1_resized_rgb)
axes[0].axis('off')
axes[0].set_title('Imagen 1')

axes[1].imshow(img2_resized_rgb)
axes[1].axis('off')
axes[1].set_title('Imagen 2')

axes[2].imshow(img3_resized_rgb)
axes[2].axis('off')
axes[2].set_title('Imagen 3')

# Mostrar la figura
plt.show()
```

```python

# Crear una nueva imagen combinando los canales
combined_img = cv2.merge((img3_resized[:, :, 0], img2_resized[:, :, 1], img1_resized[:, :, 2]))

# Convertir la imagen combinada de BGR a RGB para matplotlib
combined_img_rgb = cv2.cvtColor(combined_img, cv2.COLOR_BGR2RGB)
cv2.imwrite('combinado.jpg', combined_img_rgb)
# Mostrar la nueva imagen combinada usando matplotlib
plt.imshow(combined_img_rgb)
plt.axis('off')
plt.title('Imagen Combinada')
plt.show()
```

```python
import cv2
import numpy as np

# Cargar la imagen
image = cv2.imread('combinado.jpg')  

# Dividir la imagen en sus canales BGR
b_channel, g_channel, r_channel = cv2.split(image)

# Crear imágenes en blanco para los canales
zeros = np.zeros(image.shape[:2], dtype="uint8")

# Inicializar el estado de los canales
show_red = True
show_green = True
show_blue = True

while True:
    # Crear una imagen en blanco para mostrar los canales seleccionados
    merged_image = cv2.merge([
        b_channel if show_blue else zeros,
        g_channel if show_green else zeros,
        r_channel if show_red else zeros
    ])

    # Mostrar la imagen resultante
    cv2.imshow('Canales de Color', merged_image)

    # Esperar por una tecla y manejar la entrada del usuario
    key = cv2.waitKey(1) & 0xFF

    if key == ord('r'):  # Presiona 'r' para alternar el canal rojo
        show_red = not show_red
    elif key == ord('g'):  # Presiona 'g' para alternar el canal verde
        show_green = not show_green
    elif key == ord('b'):  # Presiona 'b' para alternar el canal azul
        show_blue = not show_blue
    elif key == 27:  # Presiona 'Esc' para salir
        break

# Cerrar todas las ventanas al finalizar
cv2.destroyAllWindows()
```
